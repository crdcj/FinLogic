```python
import os
import pandas as pd
import numpy as np

pd.options.display.max_colwidth = 15
pd.options.display.max_rows = 4
```

```python
files_names = sorted(os.listdir('./data/processed/parquet/'))
files_names
```

```python
df = pd.DataFrame()
for file_name in files_names:
    dfy = pd.read_parquet(f'./data/processed/parquet/{file_name}')
    dfy.query("CD_CVM == 9512", inplace=True)
    df = pd.concat([df, dfy], ignore_index=True)
df
```
